```python
import ollama
response = ollama.generate(model="llama3.1:8b", prompt="Why does moon glow?")
print(response)
```

```python
print(response.response)
```

```python
streamed_response = ollama.generate(model="llama3.1:8b", prompt="Why does moon glow?", stream=True)
for chunk in streamed_response:
    print(chunk["response"], end="")
```

```python
import base64

image_path = "C:\\Users\\AnkitKumar\\Pictures\\Screenshots\\Screenshot 2026-05-21 152442.png"
with open(image_path, "rb") as f:
    image_data = base64.b64encode(f.read()).decode("utf-8")

image_response = ollama.generate(model="gemma4:e4b", prompt="What is in this image?", images=[image_data])
print(image_response.response)
```

```python


images_path = [r"C:\Users\AnkitKumar\Pictures\Screenshots\Screenshot 2026-05-23 115052.png", r"C:\Users\AnkitKumar\Pictures\Screenshots\Screenshot 2026-05-23 115315.png"]
image_data_list = []

for path in images_path:
    with open(path, "rb") as f:
        image_data = base64.b64encode(f.read()).decode("utf-8")
        image_data_list.append(image_data)

multi_image_response = ollama.generate(model="gemma4:e4b", prompt="generate a story based on these images and make sure to include details from each image", images=image_data_list)
print(multi_image_response.response)
```
